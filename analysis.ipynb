```python
import pandas as pd
df = pd.read_csv("data\spotify.csv")
```

```python
df[['track_name', 'album_name']] = (
    df.groupby('artist_name')[['track_name', 'album_name']]
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))
)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=1)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
len(df.columns)
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df.dtypes
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['explicit'] = df['explicit'].astype(bool)
df['mode'] = df['mode'].map({0: 'Minor', 1: 'Major'})


```
